# Numpy

### 1

```python
def cycleLinspace(start, stop, num = 50):
    delta = (stop - start)/(num - 1)
    result = []
    
    for i in range(0, num):
        result.append(start + delta * i)
        
    return np.array(result)
```

```python
def comprehensionLinspace(start, stop, num = 50):
    delta = (stop - start)/(num - 1)
    
    return np.array([i * delta for i in range(0, num)])
```

```python
%%time
cycleLinspace(0, 10000, 10001)
```

```python
%%time
comprehensionLinspace(0, 10000, 10001)
```

```python
%%time
np.linspace(0, 10000, 10001)
```

# SciPy

```python
from scipy import optimize
```

```python
def f(x):
    return (x[0] - 3.2) ** 2 + (x[1] - 0.1) ** 2 + 3
```

```python
x_min = optimize.minimize(f, [5, 5])
print(x_min)
```

```python
print(x_min.x)
```

```python
%%time
x_min = optimize.minimize(f, [5, 5])
```

```python
x = np.arange(-10, 10, 0.1)
y = x ** 3
plt.plot(x, y, label = '$x^{3}$')
plt.title('$x^{3}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
