```python
import numpy as np
import pylab as plt
from numpy.linalg import inv, norm, eigvals, svd
from scipy import diag

'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (l, n) = F.shape
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    # U --> already transposed (U^T)
    V, d, U = svd(F, full_matrices=False)

    #print(U.shape, d.shape, V.shape, F.shape)
    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / dTot for i in range(n)])
    m = min(i for i,x in enumerate(var_i) if x >= varRetained)
    print('%.2f variance retained in %i dimensions' % (var_i[m], m + 1))

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = U[:m + 1, :]
    G = np.dot(F, np.transpose(U_reduced)) #вычислить матрицу в преобразованном пространстве

    return G, U_reduced
```

eps должен быть меньше среднего расстояния между двумя изображениями, т.к. иначе все, скорей всего, сольется в большой кластер с некоторым количеством выбросов.

```python
print("Sample size = ", y.shape[0])
```

```python
# dataframe for storing metrics of clusterization
def create_metrics_frame():
    frame_metrics = pd.DataFrame()
    frame_metrics['EPS'] = 0
    frame_metrics['Min_samples'] = 0
    frame_metrics['Homogeneity'] = 0
    frame_metrics['Completeness'] = 0
    frame_metrics['V-measure'] = 0
    frame_metrics['Adjusted Rand-Index'] = 0
    frame_metrics['Adjusted Mutual Information'] = 0
    return frame_metrics
```

```python
# clusterization and metrics calculation 
def clustering(clusterizator, X, labels, frame_metrics):
    clusterizator.fit(X)

    metrics_line = {
        'Homogeneity' : metrics.homogeneity_score(labels, clusterizator.labels_), 
        'Completeness' : metrics.completeness_score(labels, clusterizator.labels_), 
        'V-measure' : metrics.v_measure_score(labels, clusterizator.labels_), 
        'Adjusted Rand-Index' : metrics.adjusted_rand_score(labels, clusterizator.labels_), 
        'Adjusted Mutual Information' : metrics.adjusted_mutual_info_score(labels, clusterizator.labels_)}
    
    return metrics_line
```

```python
# params grid
eps = np.arange(10, 40, 1)
min_samples = np.arange(2, 40, 1)
```

```python
# clusterization with different params
frame_metrics = create_metrics_frame()
for e in eps:
    for samples in min_samples:
        clusterizator = DBSCAN(eps=e, min_samples=samples)
        metrics_line = clustering(clusterizator, X, y, frame_metrics)
        clusterizator.core_sample_indices_
        metrics_line['EPS'] = e
        metrics_line['Min_samples'] = samples
        frame_metrics.loc[len(frame_metrics.index)] = metrics_line
```

```python
# top 20 params are sorted by 'Adjusted Rand-Index'
frame_metrics.sort_values(by="Adjusted Rand-Index", ascending=False)[:20]
```

```python
# array of params and metrics for best case
best_metrics = frame_metrics.sort_values(by="Adjusted Rand-Index", ascending=False).as_matrix()[0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, MeanShift, AgglomerativeClustering
from sklearn.metrics import pairwise_distances_argmin
from sklearn.datasets import load_sample_image
from sklearn.utils import shuffle
from time import time
from skimage.measure import compare_ssim
from scipy.spatial.distance import euclidean
```

```python
def get_label(vec, centers):
    return np.argmin([euclidean(vec, c) for c in centers])

def get_cluster_centers(X, cluster_labels):
    cluster_centers = np.zeros((max(cluster_labels) + 1, X.shape[1]))
    cluster_weights = np.zeros(max(cluster_labels) + 1)
    for i in range(len(cluster_labels)):
        cluster_weights[cluster_labels[i]] += 1
        cluster_centers[cluster_labels[i]] += X[i][0]
    cluster_centers /= cluster_weights.reshape(cluster_weights.shape[0], 1)
    
    return cluster_centers
```

```python
def recreate_image(codebook, labels, w, h):
    """Recreate the (compressed) image from the code book & labels"""
    d = codebook.shape[1]
    image = np.zeros((w, h, d))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            image[i][j] = codebook[labels[label_idx]]
            label_idx += 1
    return image

def reduce_colors(image, clusterizer, algo_name, show_original=True):
    # Convert to floats instead of the default 8 bits integer coding. Dividing by
    # 255 is important so that plt.imshow behaves works well on float data (need to
    # be in the range [0-1])
    image = np.array(image, dtype=np.float64) / 255

    # Transform Image to a 2D numpy array.
    w, h, d = original_shape = tuple(image.shape)
    assert d == 3
    image_array = np.reshape(image, (w * h, d))
    
    image_array_sample = shuffle(image_array, random_state=0)[:1000]
    clusterizer.fit(image_array_sample)
    

    try:
        n_colors = len(clusterizer.cluster_centers_)
        cluster_centers = clusterizer.cluster_centers_
        # Get labels for all points
        labels = clusterizer.predict(image_array)
    except:
        n_colors = clusterizer.n_clusters
        cluster_centers = get_cluster_centers(image_array_sample, clusterizer.labels_)
        
        # Get labels for all points
        labels = [get_label(vec, cluster_centers) for vec in image_array]
        
    
    # New image
    quantized_image = recreate_image(cluster_centers, labels, w, h)

    # SSIM
    ssim = compare_ssim(image, quantized_image, multichannel=True)
    
    # Display all results, alongside original image
    if show_original:
        plt.figure(1)
        plt.clf()
        ax = plt.axes([0, 0, 1, 1])
        plt.axis('off')
        plt.title('Original image (96,615 colors)')
        plt.imshow(image)

    plt.figure(2)
    plt.clf()
    ax = plt.axes([0, 0, 1, 1])
    plt.axis('off')
    plt.title('Quantized image ('+ str(n_colors) +' colors, SSIM = ' + str(ssim) + ', ' + algo_name + ')')
    plt.imshow(quantized_image)
    plt.show()
    
    return ssim
```

```python
image = load_sample_image("china.jpg")
K = [2, 3, 10, 20]
```

```python
image_metrics = pd.DataFrame()
image_metrics['algo'] = 0
image_metrics['n_colors'] = 0
image_metrics['ssim'] = 0
metric_line = {'algo' : 0, 'n_colors' : 0, 'ssim' : 0}
```

```python
algo_name = 'K-Means'
ssims = []
for i, n_colors in enumerate(K):
    clusterizer = KMeans(n_clusters=n_colors, random_state=0)
    metric_line['algo'] = algo_name
    metric_line['n_colors'] = n_colors
    metric_line['ssim'] = reduce_colors(image, clusterizer, algo_name, show_original=(i==0))
    image_metrics = image_metrics.append(metric_line, ignore_index=True)
```

```python
algo_name = 'MeanShift'
clusterizer = MeanShift()
metric_line['algo'] = algo_name
metric_line['ssim'] = reduce_colors(image, clusterizer, algo_name, show_original=True)
metric_line['n_colors'] = len(clusterizer.cluster_centers_)
image_metrics = image_metrics.append(metric_line, ignore_index=True)
```

```python
algo_name = 'AgglomerativeClustering'
ssims = []
for i, n_colors in enumerate(K):
    clusterizer = AgglomerativeClustering(n_clusters=n_colors)
    metric_line['algo'] = algo_name
    metric_line['n_colors'] = n_colors
    metric_line['ssim'] = reduce_colors(image, clusterizer, algo_name, show_original=(i==0))
    image_metrics = image_metrics.append(metric_line, ignore_index=True)
```
