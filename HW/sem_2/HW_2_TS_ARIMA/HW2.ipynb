```python
ts_boxcox = deepcopy(np.log(train_ts + 1e-7))
ts_boxcox.plot(figsize=(15,5))
```

```python
print('optimal lambda:', stats.boxcox(train_ts + 2)[1][0])
ts_boxcox[item['item']], lmbda = stats.boxcox(train_ts + 2)
ts_boxcox.plot(figsize=(15,5))
```

### 3) Данные не стационарны => продифференцируем

```python
diff_ts_boxcox = deepcopy((ts_boxcox  - ts_boxcox.shift(1))[1:])
diff_ts_boxcox.plot(figsize=(15,6))
```

```python
# Augmented Dickey-Fuller unit root test
df_stat = stattools.adfuller(np.array(diff_ts_boxcox[item['item']].dropna()))
print('The Augmented Dickey-Fuller test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ncritical values: %s' % 
      (df_stat[0], df_stat[1], df_stat[2], df_stat[4]))
```

Из этого можно сделать вывод, что данные стационарны.

### 4) Определим p и q по ACF/PACF

```python
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(diff_ts_boxcox.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(diff_ts_boxcox.values.squeeze(), lags=40, ax=ax2)
```

### 5) Здадим набор возможных p и q для ARMA

```python
q_set = np.arange(0, 4, 1)
p_set = np.arange(0, 11, 1)
d_set = [1]
```

### 6) Затреним ARIMA и выберем по AIC лучшие параметры

```python
%%time
aic_score = dict()
## Choosing best model accoridng to AIC criterion
for p in p_set:
    for d in d_set:
        for q in q_set:
            try:
                arima_mod = sm.tsa.ARIMA(diff_ts_boxcox, (p, d, q)).fit()
                aic_score[(p, d, q)] = arima_mod.aic
            except:
                aic_score[(p, d, q)] = None
```

```python
aic_score
```

```python
for_min_params = {key : val 
                  for key, val in aic_score.items() if val is not None}
best_params = min(for_min_params, key=for_min_params.get)
```

```python
best_params
```
